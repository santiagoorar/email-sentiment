The dataset has only 5842 observations, there are too few observations to train a deep learning model. ThatÂ´s why a pretrained model from hugging face will be used to solve this.

```python
# LetÂ´s first plot the distribution of sentiments
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Sentiment')
plt.title('Distribution of Sentiments')
plt.show()
```

The dataset is imbalanced with a higher number of observations of neutral sentiments. This is important to remark because when training the model, this imbalance might lead the model to be more biased towards predicting neutral sentiments. But still does not look as resampling techniques to increase the other classes would be necessary.

```python
# Word count distribution
df['word_count'] = df['Sentence'].apply(lambda x: len(str(x).split()))
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='word_count', bins=50, kde=True)
plt.title('Word Count Distribution')
plt.show()
```

The distribution looks to be right-skewed. Most sentences in the dataset have a word count that falls between approximately 5 and 30 words.

Looking at the distribution we can think about its implications in the model training and in the selection of the pre-trained model to use. Given this distribution, a reasonable maximum sequence length for tokenizing during training might be around 50 to 60 words to accommodate most sentences without truncating much information.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Sentiment', y='word_count')
plt.title('Word Count Distribution by Sentiment')
plt.show()
```

We can see that the lenght of the text is similar despite the sentiment. 

Due the varying lenghts the use of Transformers (because of the attention mechanisms) probably will be good. 

```python
# Now, letÂ´s look at the most frequent words per sentiment:

stopwords = set(STOPWORDS)


for sentiment in ['positive', 'negative', 'neutral']:
    plt.figure(figsize=(10, 5))
    wc = WordCloud(background_color='white', stopwords=stopwords, max_words=200, max_font_size=40, random_state=42)
    wc.generate(' '.join(df[df['Sentiment'] == sentiment]['Sentence']))
    plt.imshow(wc, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'Most Frequent Words in {sentiment.capitalize()} Sentiments')
    plt.show()
```
